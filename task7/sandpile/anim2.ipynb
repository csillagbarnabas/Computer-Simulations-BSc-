```python
%pylab inline
import matplotlib.animation as animation
import matplotlib.pyplot as plt

fig = plt.figure( figsize=(8,8) )

data = loadtxt("sandpileflow.dat")

a = data[0,:].reshape(100,100)
im = plt.imshow(a, aspect='auto', vmin = 0, vmax=7)

def animate_func(i):
    im.set_array(data[i,:].reshape(100,100))
    return [im]

anim = animation.FuncAnimation(
                               fig, 
                               animate_func, 
                               frames = 199,
                               interval = 10,
                               )
anim.save('sandpile_flow.gif')

print('Done!')
```
