```python
%pylab inline
import matplotlib.animation as animation
import matplotlib.pyplot as plt

fig = plt.figure( figsize=(8,8) )

data = loadtxt("conwaypern2h.dat")*100

a = data[0,:].reshape(200,200)
im = plt.imshow(a, interpolation='none', aspect='auto', vmin=0, vmax=1)

def animate_func(i):
    im.set_array(data[i,:].reshape(200,200))
    return [im]

anim = animation.FuncAnimation(
                               fig, 
                               animate_func, 
                               frames = 399,
                               interval = 10,
                               )

anim.save('perhatarn2h.gif')

print('Done!')
```

```python
dat1 = loadtxt("conwayelon1.dat")
dat2 = loadtxt("conwaypern1.dat")
dat3 = loadtxt("conwaykonstn1.dat")
figsize(18,6)
subplot(1,3,1)
plt.imshow(dat1[99,:].reshape(200,200)*100, interpolation='none', aspect='auto', vmin=0, vmax=1)
plt.title("Élő határ",fontsize=12)
subplot(1,3,2)
plt.imshow(dat2[99,:].reshape(200,200)*100, interpolation='none', aspect='auto', vmin=0, vmax=1)
plt.title("Periodikus határfeltétel",fontsize=12)
subplot(1,3,3)
plt.imshow(dat3[99,:].reshape(200,200)*100, interpolation='none', aspect='auto', vmin=0, vmax=1)
plt.title("Konstans véletlen határ",fontsize=12)
```
